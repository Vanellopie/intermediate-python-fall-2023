```python
import pandas as pd
from ds_utils.preprocessing.clean_data import remove_duplicates, handle_missing_values
from ds_utils.preprocessing.feature_engineering import normalize_data
from ds_utils.visualization.plot_data import scatter_plot

# Create a sample DataFrame
data = pd.DataFrame({'A': [1, 2, 2, 3, 4], 'B': [5, 6, 7, None, 9]})

# feature_engineering.py

def normalize_data(data):
    normalized_data = (data - data.mean()) / data.std()
    
    return normalized_data

# Data preprocessing
cleaned_data = remove_duplicates(data)
cleaned_data = handle_missing_values(cleaned_data)
normalized_data = normalize_data(cleaned_data)

# Data visualization
scatter_plot(normalized_data['A'], normalized_data['B'])
```
