```python
response = requests.get('https://www.zangia.mn/job/list/pg.1/b.4')
soup = BeautifulSoup(response.content)

ads = soup.find_all("div", {"class": "ad"})
ads[0].find('b').get_text()

job_titles = []
for ad in ads:
    job_title = ad.find('b').get_text()
    job_titles.append(job_title)

job_titles
```

```python
job_salary = soup.find_all("span", {"class":"fsal"})
job_salary = [x.get_text().strip() for x in job_salary]

job_salary
```

```python
df = pd.DataFrame({"Job Titles": job_titles, "Salary": job_salary})
df
```

```python
for i in range(len(df)):
    if "Тохиролцоно" in df['Salary'][i]:
        df = df.drop(i)
df
```
